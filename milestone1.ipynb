## Start Looking for better ways to show qualitative data in your dataset


```python
mx_data.plot(x="Year",y=["Agricultural land (Area)","Arable land (Area)","Renewable energy consumption (consumption)","Access to electricity ( Population)"])
```

```python
mx_data.plot(x="Year",y=["CO2 emissions from gaseous fuel consumption (kt)","CO2 emissions (kt)","Other greenhouse gas emissions, HFC, PFC and SF6 (thousand metric tons of CO2 equivalent)","Total greenhouse gas emissions (kt of CO2 equivalent)"])
```

```python
sns.boxplot(x=mx_data["Agricultural land (Area)"])
```

```python
corr = mx_data.corr()
display(corr)
```

```python
sns.heatmap(
    corr,
    annot=True,
)
```
